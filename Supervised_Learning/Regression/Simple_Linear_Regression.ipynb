```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Import ML Libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('C:\\Users\\Navneet\\Documents\\Machine Learning Nov\\Datasets\\carsspeed.csv', index_col=0)
df.head()
```

```python
df.plot(kind='scatter', x = 'speed', y = 'dist', figsize=(10, 6), color = 'r', s = 50)
```

```python
#Simple Linear Regression 
#
X = df[['speed']]
y = df[['dist']]
reg = LinearRegression()
reg.fit(X.values, y.values) #train the algorithm
print(f'slope: {reg.coef_}')
```

```python
y_predict = reg.predict(X.values)
```

```python
print(f"MSE = {mean_squared_error(y, y_predict)}")
print(f"Accuracy = {r2_score(y, y_predict)*100}")
```

```python
df.plot(kind='scatter', x='speed', y='dist', figsize=(10, 6), color='r', s=50)

plt.plot(X.values, y_predict, color = 'b')
plt.show()
```
